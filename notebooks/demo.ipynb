```python
# 1. (Opcional) Faça o download dos pesos do modelo caso ainda não tenha feito
from src.downloader import ModelDownloader

downloader = ModelDownloader()
downloader.run_all_downloads()
```

```python
# 2. Gere as máscaras de movimento usando DensePose
from src.pose_extractor import PoseExtractor

extractor = PoseExtractor()
masked_video_path = extractor.extract_poses("caminho/para/seu/video.mp4")
print(f"Máscara de vídeo gerada em: {masked_video_path}")
```

```python
# 3. Anime a sua imagem de referência baseada na máscara de movimento
from src.animator import Animator

animator = Animator()
animation_path = animator.animate(
    reference_image_path="caminho/para/sua/imagem.jpg",
    motion_sequence_path=masked_video_path,
    seed=2,
    steps=35,
    guidance_scale=7.5
)

print(f"Vídeo gerado salvo em: {animation_path}")
```
